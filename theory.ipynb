### use fuzzy words to get similar tags replaced

```python
from fuzzywuzzy import fuzz
import string
# Define a function to find the most similar game category

def find_similar_category(category, category_list):
    max_ratio = 0
    similar_category = None

    translator = str.maketrans('', '', string.punctuation) 

    for item in category_list:
        item_clean = item.translate(translator).replace(' ', '')
        category_clean = category.translate(translator).replace(' ', '')

        ratio = fuzz.ratio(category_clean.lower(), item_clean.lower())
        if ratio > max_ratio:
            max_ratio = ratio
            similar_category = item_clean

    return similar_category


for i, category in enumerate(DeletedRowData['Game Category']):
    similar_category = find_similar_category(category, DeletedRowData['Game Category'])
    DeletedRowData['Game Category'].replace(category,similar_category,inplace=True ) 
    print(category+" is replaced by "+ similar_category)
```

# Preprocessing and BOW for Scripts


```python
from nltk.tokenize import RegexpTokenizer # this will remove everything that's not alphanumeric characters
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords
import nltk
from sklearn.feature_extraction.text import CountVectorizer
from nltk.stem import PorterStemmer

#nltk.download('stopwords')
#nltk.download('wordnet')
#nltk.download('punkt')
def preprocess(stopwords, script):
    # lower case
    script=script.lower()
    # unicode removal
    script = script.encode(encoding="ascii", errors="ignore")# decoding the text
    script = script.decode()
    #tocken
    tokenizer = RegexpTokenizer(r'[a-zA-Z]{2,}') # removing less than 2 characters
    tokens=tokenizer.tokenize(script)
    # remove words that are made of only repeated letters and stop words
    token= [ words for words in tokens if words not in stopwords and not any(char * len(words) == words for char in words)] 
    # remove  3 or more repeated characters consecutively 
    token = [word for word in token if not re.search(r'(.)\1{2,}', word)]
    # lemma
    lemmatizer = WordNetLemmatizer()
    token= [lemmatizer.lemmatize(word) for word in token]
    # stem
    Ps=PorterStemmer()
    token= [Ps.stem(word) for word in token]
    # keep only words that have 3 or more characters
    token = [word for word in token if len(word) > 2]
    script= " ".join(token)
    return script

stopwords = set(stopwords.words('english'))
words=[]
for i in DeletedRowData['Scripts']:
    words.append(preprocess(stopwords,i))
vectorizer = CountVectorizer()
Bow = vectorizer.fit_transform(words)
feature_names = vectorizer.get_feature_names_out()
pd.DataFrame(Bow.toarray(),columns= feature_names)
```

# IF-IDF

```python
from sklearn.feature_extraction.text import TfidfTransformer

transformer = TfidfTransformer()
tfidf = transformer.fit_transform(Bow)
tfidf_array = tfidf.toarray()

# Create a DataFrame to display the words and their TF-IDF values
tfidf_df = pd.DataFrame(tfidf_array, columns=feature_names)
tfidf_df
```

```python
# Calculate the sum of TF-IDF values across all documents
total_tfidf = tfidf_df.sum(axis=0)
top_words = total_tfidf.nlargest(150)
plt.figure(figsize=(30, 10))
plt.bar(top_words.index, top_words.values, color='skyblue')
plt.title('Top 150 Most Used Words')
plt.xlabel('Words')
plt.ylabel('TF-IDF Value')
plt.xticks(rotation=80)
plt.show()
```
