```python
DeletedRowData
import unicodedata
# remove spaces and accented characters
def remove_accents(text):
    return ''.join(c for c in unicodedata.normalize('NFD', text) if unicodedata.category(c) != 'Mn')

DeletedRowData['Game Category'] = DeletedRowData['Game Category'].str.lower().str.replace(' ', '').apply(remove_accents)

```

### use fuzzy words to get similar tags replaced

```python
from fuzzywuzzy import fuzz
import string
# Define a function to find the most similar game category

def find_similar_category(category, category_list):
    max_ratio = 0
    similar_category = None

    translator = str.maketrans('', '', string.punctuation) 

    for item in category_list:
        item_clean = item.translate(translator).replace(' ', '')
        category_clean = category.translate(translator).replace(' ', '')

        ratio = fuzz.ratio(category_clean.lower(), item_clean.lower())
        if ratio > max_ratio:
            max_ratio = ratio
            similar_category = item_clean

    return similar_category


for i, category in enumerate(DeletedRowData['Game Category']):
    similar_category = find_similar_category(category, DeletedRowData['Game Category'])
    DeletedRowData['Game Category'].replace(category,similar_category,inplace=True ) 
    print(category+" is replaced by "+ similar_category)
```
